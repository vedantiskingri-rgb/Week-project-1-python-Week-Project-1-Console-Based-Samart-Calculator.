# Week Project 1: Console-Based Samart Calculator.

```python
import numpy as np

def show_menu():
    print("\n--- Smart Calculator ---")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Power")
    print("6. Modulus")
    print("7. Exit")

while True:
    show_menu()
    choice = int(input("Enter your choice (1-7): "))

    if choice == 7:
        print("Calculator Closed. Thank you!")
        break

    n = int(input("How many numbers do you want to use? "))

    numbers = []
    for i in range(n):
        num = float(input(f"Enter number {i+1}: "))
        numbers.append(num)

    # Convert list to NumPy array
    arr = np.array(numbers)

    if choice == 1:
        result = np.sum(arr)
        print("Result (Addition):", result)

    elif choice == 2:
        result = arr[0]
        for i in arr[1:]:
            result -= i
        print("Result (Subtraction):", result)

    elif choice == 3:
        result = np.prod(arr)
        print("Result (Multiplication):", result)

    elif choice == 4:
        result = arr[0]
        try:
            for i in arr[1:]:
                if i == 0:
                    raise ZeroDivisionError
                result /= i
            print("Result (Division):", result)
        except ZeroDivisionError:
            print("Error: Division by zero is not allowed")

    elif choice == 5:
        if n == 2:
            result = np.power(arr[0], arr[1])
            print("Result (Power):", result)
        else:
            print("Power operation needs exactly 2 numbers")

    elif choice == 6:
        if n == 2:
            result = arr[0] % arr[1]
            print("Result (Modulus):", result)
        else:
            print("Modulus operation needs exactly 2 numbers")

    else:
        print("Invalid choice! Please select again.")

    cont = input("Do you want to continue? (yes/no): ")
    if cont.lower() != "yes":
        print("Goodbye!")
        break
```


*   Smart Calculator are Multiple Times use.
*    Do you want to continue to command you can say yes or no.

*   The user Selects an Operation by entering a Choice number.
*   Input Values are stored and Procesed using a NumPy array.





# Week Project 2: Student Marks Analyzer  

```python
import numpy as np

n = int(input("Enter number of students: "))

marks = []

for i in range(n):
    m = float(input(f"Enter marks of student {i+1}: "))
    marks.append(m)

# Convert list to NumPy array
marks_array = np.array(marks)

total = np.sum(marks_array)
average = np.mean(marks_array)
highest = np.max(marks_array)
lowest = np.min(marks_array)

# Grade calculation
if average > 85:
    grade = "A"
elif average >= 70:
    grade = "B"
elif average >= 50:
    grade = "C"
else:
    grade = "Fail"

print("\n--- Student Marks Analysis ---")
print("Total Marks:", total)
print("Average Marks:", average)
print("Highest Marks:", highest)
print("Lowest Marks:", lowest)
print("Grade:", grade)
```


*   The program user to  enter the number of Student.

*   All entered marks are Stored in aNumPy array.


*   The Calculates & Displays the total marks.
